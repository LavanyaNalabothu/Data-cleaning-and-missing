```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("test.csv")  # replace with your path if needed

# Step 2: Analyze dataset BEFORE cleaning
print("=== BEFORE CLEANING ===")
print("Dataset shape (rows, columns):", df.shape)
print("\nMissing values per column:\n", df.isnull().sum())
print("\nPercentage of missing values per column:\n", (df.isnull().mean()*100))
print("\nData types:\n", df.dtypes)
print("\nStatistical summary:\n", df.describe())
print("\nPreview of data:\n", df.head())

# Step 3: Remove columns with extremely high missing values (threshold: 50%)
threshold = 50
missing_percent = df.isnull().mean() * 100
cols_to_drop = missing_percent[missing_percent > threshold].index
df_cleaned = df.drop(columns=cols_to_drop)

print(f"\nColumns dropped due to > {threshold}% missing values:\n", cols_to_drop)

# Step 4: Validate dataset AFTER cleaning
print("\n=== AFTER CLEANING ===")
print("Dataset shape (rows, columns):", df_cleaned.shape)
print("\nRemaining missing values per column:\n", df_cleaned.isnull().sum())
print("\nPercentage of remaining missing values per column:\n", (df_cleaned.isnull().mean()*100))
print("\nData types:\n", df_cleaned.dtypes)
print("\nStatistical summary:\n", df_cleaned.describe())
print("\nPreview of cleaned data:\n", df_cleaned.head())

# Step 5: Compare before vs after dataset size and quality
print("\n=== COMPARISON ===")
print(f"Number of columns before cleaning: {df.shape[1]}")
print(f"Number of columns after cleaning: {df_cleaned.shape[1]}")
print(f"Total missing values before cleaning: {df.isnull().sum().sum()}")
print(f"Total missing values after cleaning: {df_cleaned.isnull().sum().sum()}")

# Optional: Save cleaned dataset to a new CSV
df_cleaned.to_csv("test_cleaned.csv", index=False)
print("\nCleaned dataset saved as 'test_cleaned.csv'")
```

## Medical Appointment No Shows

```python
df2=pd.read_csv("MedicalAppointment.csv")
missing_values = df2.isnull().sum()
missing_values
```

```python
import matplotlib.pyplot as plt

# Calculate missing values
missing = df2.isnull().sum()

# Keep only columns with missing values
missing = missing[missing > 0]

# Check if missing values exist before plotting
if not missing.empty:
    plt.figure(figsize=(10,5))
    missing.plot(kind='bar')
    plt.title("Missing Values per Column")
    plt.xlabel("Columns")
    plt.ylabel("Number of Missing Values")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No missing values found in the dataset.")

```

```python
# Identify numerical columns
num_cols = df2.select_dtypes(include=np.number).columns

# Apply median imputation
df2[num_cols] = df2[num_cols].fillna(df2[num_cols].median())

# Verify numerical columns
df2[num_cols].isnull().sum()
```

```python
# Identify categorical columns
cat_cols = df2.select_dtypes(exclude=np.number).columns

# Apply mode imputation safely
for col in cat_cols:
    mode_value = df2[col].mode().iloc[0]
    df2[col] = df2[col].fillna(mode_value)

# Verify categorical columns
df2[cat_cols].isnull().sum()
```

```python
print("Total missing values after imputation:", df2.isnull().sum().sum())
```

```python
# Calculate missing value percentages for each column
missing_percent = df2.isnull().mean() * 100
print("Missing values percentage per column:\n", missing_percent)

# Remove columns with more than 50% missing values
threshold = 50
cols_to_drop = missing_percent[missing_percent > threshold].index
df_cleaned = df2.drop(columns=cols_to_drop)

print(f"\nColumns dropped due to high missing values (> {threshold}%):\n", cols_to_drop)
```

```python
# Check remaining missing values
print("\nRemaining missing values:\n", df_cleaned.isnull().sum())

# Check dataset info (data types and non-null counts)
print("\nDataset info after cleaning:")
print(df_cleaned.info())

# Quick statistical summary for numerical columns
print("\nStatistical summary:")
print(df_cleaned.describe())

# Optional: Check first few rows to ensure data looks okay
print("\nPreview of cleaned dataset:")
print(df_cleaned.head())
```
